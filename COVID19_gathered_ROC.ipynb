```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_dnn = pd.read_csv("C:\workspace\project_data\lata\comp\clf_roc\clf_LRdnn.csv")
df_skl = pd.read_csv("C:\workspace\project_data\lata\comp\clf_roc\clf_LRskl.csv")
df_rnd = pd.read_csv("C:\workspace\project_data\lata\comp\clf_roc\clf_randomforest.csv")
df_xgb = pd.read_csv("C:\workspace\project_data\lata\comp\clf_roc\clf_xgboost.csv")
```

```python
x1dnn = df_dnn['fpr']
y1dnn = df_dnn['tpr']
```

```python
x2skl = df_skl['fpr']
y2skl = df_skl['tpr']
```

```python
x3rnd = df_rnd['fpr']
y3rnd = df_rnd['tpr']
```

```python
x4xbg = df_xgb['fpr']
y4xbg = df_xgb['tpr']
```

```python
plt.figure(figsize=(8,8))
#plt.rc('font', size=10)
plt.rc('axes', titlesize=25)   
plt.rc('axes', labelsize=20)  
plt.rc('xtick', labelsize=15)   
plt.rc('ytick', labelsize=15) 
plt.rc('legend', fontsize=13)
#plt.rc('figure', titlesize=20)

plt.plot(x1dnn, y1dnn, ':', linewidth=4.5, label = "DNN")
plt.plot(x2skl, y2skl, '--', linewidth=4.5, label = "sklearn_logistic")
plt.plot(x3rnd, y3rnd, '-.', linewidth=4.5, label = "Random_Forest")
plt.plot(x4xbg, y4xbg, linewidth=4.5, label = "XGBoost")
plt.plot([0, 1], [0, 1], 'k--', linewidth=2, label="random guess")
plt.xlabel('False positive rate')
plt.ylabel('True positive rate')
plt.title('Classification')
plt.legend()

plt.savefig('roc.png', dpi=300)
```

```python
plt.figure(figsize=(12,12))
#plt.rc('font', size=10)
plt.rc('axes', titlesize=20)   
plt.rc('axes', labelsize=15)  
plt.rc('xtick', labelsize=15)   
plt.rc('ytick', labelsize=15) 
plt.rc('legend', fontsize=12)
#plt.rc('figure', titlesize=20)



plt.subplot(2,2,1)
plt.tight_layout(pad=8.0)
plt.plot(x1dnn, y1dnn, ':', color='b',linewidth=3, label = "DNN")
plt.plot([0, 1], [0, 1], 'k--',linewidth=2)
plt.xlabel('False positive rate')
plt.ylabel('True positive rate')
plt.title('Classification - DNN')
plt.legend()

plt.subplot(2,2,2)
plt.plot(x2skl, y2skl, '--', color='orange',linewidth=3, label = "sklearn_logistic")
plt.plot([0, 1], [0, 1], 'k--',linewidth=2)
plt.xlabel('False positive rate')
plt.ylabel('True positive rate')
plt.title('Classification - skl_regression')
plt.legend()

plt.subplot(2,2,3)
plt.plot(x3rnd, y3rnd, '-.', color='g',linewidth=3, label = "Random_Forest")
plt.plot([0, 1], [0, 1], 'k--',linewidth=2)
plt.xlabel('False positive rate')
plt.ylabel('True positive rate')
plt.title('Classification - rand_forest')
plt.legend()

plt.subplot(2,2,4)
plt.plot(x4xbg, y4xbg, color='r',linewidth=3, label = "XGBoost")
plt.plot([0, 1], [0, 1], 'k--',linewidth=2)
plt.xlabel('False positive rate')
plt.ylabel('True positive rate')
plt.title('Classification - XGBoost')
plt.legend()

plt.savefig('sub_roc.png', dpi=300)
```
